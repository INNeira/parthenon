```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Binary interposition with `tau_exec` was used to collect profiling information for runs with nGPUs=1-32. The environment variables `TAU_TRACK_MESSAGE`, `TAU_TRACK_MESSAGE`, `TAU_PROFILE`, and `TAU_TRACE` were all set to 1 (one), i.e. enabled. The `profile.*` files generated for each run were combined into summary text files with `pprof -sp > <ngpus>.txt`. 

Each summary file contains profile data for function calls both as "total" time (total time spent in a function per run) and as "mean" time (total time per process (i.e. effectively per GPU) per run).

Read each section of the summary files into pandas DataFrames for easy processing/plotting. Add the number of GPUs for each run in a column. Verify I got all of the data by printing the first and last rows.

Collect all the per-run data into joint dataframes. Use the `MultiIndex` feature to store identifying data (function name, nGPUs).
And display what we have so far.

```python
total_df = run1_total_df.set_index(['Name', 'ngpus'])
total_df = total_df.append(run2_total_df.set_index(['Name', 'ngpus']))
total_df = total_df.append(run4_total_df.set_index(['Name', 'ngpus']))
total_df = total_df.append(run8_total_df.set_index(['Name', 'ngpus']))
total_df = total_df.append(run16_total_df.set_index(['Name', 'ngpus']))
total_df = total_df.append(run32_total_df.set_index(['Name', 'ngpus']))
total_df
```

```python
mean_df = run1_mean_df.set_index(['Name', 'ngpus'])
mean_df = mean_df.append(run2_mean_df.set_index(['Name', 'ngpus']))
mean_df = mean_df.append(run4_mean_df.set_index(['Name', 'ngpus']))
mean_df = mean_df.append(run8_mean_df.set_index(['Name', 'ngpus']))
mean_df = mean_df.append(run16_mean_df.set_index(['Name', 'ngpus']))
mean_df = mean_df.append(run32_mean_df.set_index(['Name', 'ngpus']))
mean_df
```

Create lists of the top `topN` Kokkos- and MPI- functions each that had the highest value in the 'Excl. msec' `column`, and separately for the "total" and "mean" values. I'm using the data for the run with 32 GPUs to find the rankings. These are used later to filter the data for plotting.

```python
topN = 10
column = 'Excl. msec'
```

```python
top_total_kokkos = run32_total_df[run32_total_df['Name'].str.startswith("Kokkos")]\
                .sort_values(column, ascending=False).head(n=topN)['Name']
top_total_kokkos.to_list()
```

```python
top_total_mpi = run32_total_df[run32_total_df['Name'].str.startswith("MPI")]\
            .sort_values(column, ascending=False).head(n=topN)['Name']
top_total_mpi.to_list()
```

```python
top_mean_kokkos = run32_mean_df[run32_mean_df['Name'].str.startswith("Kokkos")]\
                .sort_values(column, ascending=False).head(n=topN)['Name']
top_mean_kokkos.to_list()
```

```python
top_mean_mpi = run32_mean_df[run32_mean_df['Name'].str.startswith("MPI")]\
                .sort_values(column, ascending=False).head(n=topN)['Name']
top_mean_mpi.to_list()
```

Let's make a bunch of plots, filtered by the `topN` lists I made earlier and the column of interest. 

stackoverflow tells me the magic for plotting multiIndex DataFrames correctly/nicely is `.unstack(level=0)`.

```python
# Plot 'Excl. msec' for Kokkos and MPI functions
col_name = 'Excl. msec'
```

```python
ax0 = total_df.loc[top_total_kokkos.to_list(),col_name].unstack(level=0)\
            .plot(marker='o', title="Weakscaling on rzAnsel (Total)", logx=True)
ax0.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax0.set_ylabel(col_name)
```

```python
ax1 = total_df.loc[top_total_mpi.to_list(), col_name].unstack(level=0)\
             .plot(marker='o', title="Weakscaling on rzAnsel (Total)", logx=True)
ax1.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax1.set_ylabel(col_name)
```

```python
ax2 = mean_df.loc[top_mean_kokkos.to_list(),col_name].unstack(level=0)\
            .plot(marker='o', title="Weakscaling on rzAnsel (Total/GPU)", logx=True)
ax2.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax2.set_ylabel(col_name)
```

```python
ax3 = mean_df.loc[top_mean_mpi.to_list(), col_name].unstack(level=0)\
            .plot(marker='o', title="Weakscaling on rzAnsel (Total/GPU)", logx=True)
ax3.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax3.set_ylabel(col_name)
```

```python
# Plot 'Incl. usec/call' for Kokkos and MPI functions, which is the same for 'Total' and 
# 'Mean' quantities since it's normalized by the number of function calls.
# Note that time units are now usec (Tau's decision)
col_name = 'Incl. usec/call'
```

```python
ax4 = total_df.loc[top_total_kokkos.to_list(),col_name].unstack(level=0)\
            .plot(marker='o', title="Weakscaling on rzAnsel (per call)", logx=True)
ax4.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax4.set_ylabel(col_name)
```

```python
ax5 = total_df.loc[top_total_mpi.to_list(), col_name].unstack(level=0)\
            .plot(marker='o', title="Weakscaling on rzAnsel (per call)", logx=True)
ax5.legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
ax5.set_ylabel(col_name)
```
